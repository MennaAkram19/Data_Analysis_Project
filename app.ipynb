```python
%%writefile app.py
import streamlit as st
import pandas as pd
import joblib

# ---------------- Load model ----------------
model = joblib.load("churn_model.pkl")
model_features = model.feature_names_in_

st.title("üìû Telco Churn Prediction")
st.write("Enter customer information to predict churn.")
# ---------------- User Inputs ----------------
tenure = st.number_input("Tenure (months)", 0, 100, 12)
monthly = st.number_input("Monthly Charges", 0.0, 200.0, 70.0)
total = st.number_input("Total Charges", 0.0, 10000.0, 800.0)

gender = st.selectbox("Gender", ["Female", "Male"])
contract = st.selectbox("Contract Type", ["Month-to-month", "One year", "Two year"])
internet = st.selectbox("Internet Service", ["DSL", "Fiber optic", "No"])
dependents = st.selectbox("Dependents", ["Yes", "No"])
tech = st.selectbox("Tech Support", ["Yes", "No"])
multiple = st.selectbox("Multiple Lines", ["Yes", "No"])
online_security = st.selectbox("Online Security", ["Yes", "No"])
online_backup = st.selectbox("Online Backup", ["Yes", "No"])
device_protection = st.selectbox("Device Protection", ["Yes", "No"])
streaming_tv = st.selectbox("Streaming TV", ["Yes", "No"])
streaming_movies = st.selectbox("Streaming Movies", ["Yes", "No"])
paperless = st.selectbox("Paperless Billing", ["Yes", "No"])
payment = st.selectbox("Payment Method", ["Electronic check","Mailed check","Bank transfer (automatic)","Credit card (automatic)"])

# ---------------- Create input dataframe ----------------
input_df = pd.DataFrame(0, index=[0], columns=model_features)

# Fill numeric values
input_df["tenure"] = tenure
input_df["MonthlyCharges"] = monthly
input_df["TotalCharges"] = total

# Encode categorical features
def encode(column_name, value):
    if column_name in input_df.columns:
        input_df[column_name] = value

encode("gender_Male", 1 if gender=="Male" else 0)
encode("Contract_One year", 1 if contract=="One year" else 0)
encode("Contract_Two year", 1 if contract=="Two year" else 0)
encode("InternetService_DSL", 1 if internet=="DSL" else 0)
encode("InternetService_Fiber optic", 1 if internet=="Fiber optic" else 0)
encode("InternetService_No", 1 if internet=="No" else 0)
encode("Dependents_Yes", 1 if dependents=="Yes" else 0)
encode("TechSupport_Yes", 1 if tech=="Yes" else 0)
encode("MultipleLines_Yes", 1 if multiple=="Yes" else 0)
encode("OnlineSecurity_Yes", 1 if online_security=="Yes" else 0)
encode("OnlineBackup_Yes", 1 if online_backup=="Yes" else 0)
encode("DeviceProtection_Yes", 1 if device_protection=="Yes" else 0)
encode("StreamingTV_Yes", 1 if streaming_tv=="Yes" else 0)
encode("StreamingMovies_Yes", 1 if streaming_movies=="Yes" else 0)
encode("PaperlessBilling_Yes", 1 if paperless=="Yes" else 0)
encode("PaymentMethod_Electronic check", 1 if payment=="Electronic check" else 0)
encode("PaymentMethod_Mailed check", 1 if payment=="Mailed check" else 0)
encode("PaymentMethod_Bank transfer (automatic)", 1 if payment=="Bank transfer (automatic)" else 0)
encode("PaymentMethod_Credit card (automatic)", 1 if payment=="Credit card (automatic)" else 0)

# ---------------- Prediction ----------------
if st.button("Predict Churn"):
    prob = model.predict_proba(input_df)[0][1]
    st.subheader("Prediction Result")
    st.write(f"üîÆ Churn Probability: **{prob:.2f}**")

    if prob >= 0.4:
        st.error("‚ö†Ô∏è Medium risk of churn")
    elif prob >= 0.3:
        st.warning("‚ö†Ô∏è High risk of churn")
    else:
        st.success("‚úÖ Customer likely to stay")
```
